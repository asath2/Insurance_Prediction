```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
%matplotlib inline
```

```python
df=pd.read_excel("data.xlsx")
df.head()
```

```python
plt.scatter(df.Age,df.Buy,marker=".")
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(df[['Age']],df.Buy,train_size=0.9)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
joblib.dump(model, 'model.pkl')
```
